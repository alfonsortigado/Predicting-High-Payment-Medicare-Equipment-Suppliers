# Predicting High-Payment Medicare Equipment Suppliers 

## Introduction

This notebook explores data from the **Medicare Durable Medical Equipment, Devices & Supplies by Supplier** dataset, published by the **Centers for Medicare & Medicaid Services (CMS)**. The dataset summarizes billing and beneficiary information by supplier National Provider Identifier (NPI) for services provided under Medicare Part B.

The data comes from the CMS Chronic Conditions Data Warehouse (CCW), which includes 100% of final-action claims for non-institutional Durable Medical Equipment, Prosthetics, Orthotics, and Supplies (DMEPOS) items.

---

### Dataset Overview

The dataset contains over 90 variables for more than 60,000 suppliers. These include:

- **Supplier information**: location, specialty, and type of ownership
- **Service volume**: number of claims, beneficiaries, and HCPCS codes billed
- **Medicare payment metrics**: submitted charges, allowed amounts, and standardized payments
- **Beneficiary demographics**: average age, gender counts, race breakdown
- **Chronic condition prevalence**: percent of patients with diabetes, COPD, hypertension, and more
    
URL: https://data.cms.gov/provider-summary-by-type-of-service/medicare-durable-medical-equipment-devices-supplies/medicare-durable-medical-equipment-devices-supplies-by-supplier

---

### Objective

The goal of this notebook is to build a **predictive model** that identifies whether a supplier belongs to the **top 25% of Medicare standardized payment recipients**. This is framed as a **binary classification problem**, where:

- `1` = high-payment supplier (above the 75th percentile)
- `0` = all others

---

### Why This Matters

Understanding the characteristics of high-payment suppliers can help:

- **Monitor billing patterns**
- **Flag potential anomalies or outliers**
- **Support policy analysis for cost optimization**
- **Aid audits or regulatory reviews**

By applying machine learning to CMS data, we can extract insights that complement traditional analytics, enabling faster and more scalable understanding of provider behaviors in the U.S. healthcare system.



## 1. Importing Required Libraries
This cell imports the necessary Python libraries used throughout the notebook. These libraries include:

- `pandas`: for data loading, inspection, and manipulation
- `scikit-learn` modules: for model training (`RandomForestClassifier`), evaluation (`classification_report`, `confusion_matrix`, `roc_auc_score`), and splitting the dataset into training and testing subsets
- `matplotlib.pyplot`: for visualizing feature importances and evaluation results


```python
# Core Libraries for Data Handling and Visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import copy

# Preprocessing
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Model Selection and Evaluation
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, classification_report, confusion_matrix
)

# Classification Algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

```

## 2. Load and Preview the Dataset

This cell loads the dataset into a DataFrame using `pandas.read_csv()`. The dataset contains information about Medicare suppliers of durable medical equipment, including billing amounts, service volumes, and patient characteristics.

After loading, it displays the shape of the dataset to confirm the number of records and features. The `.head()` method is used to preview the first five rows, helping to verify that the data has been loaded correctly and to provide an overview of the column structure.


```python

# Load the CSV file (replace with actual file name if needed)
df = pd.read_csv("mup_dme_ry25_p05_v10_dy23_supr.csv")  

# Show dimensions and preview
print("Dataset shape:", df.shape)
df.head()
```

## 3 Define Binary Target Variable (`high_payment_provider`)
This cell prepares the target variable for the classification model.

- First, the `Suplr_Mdcr_Stdzd_Pymt_Amt` column, which represents the standardized Medicare payment amount per supplier, is explicitly converted to numeric format. This ensures compatibility with downstream calculations and handles any potential formatting inconsistencies.

- Then, rows with missing values in this column are dropped, since they would interfere with the calculation of quantiles and target assignment.

- A new binary target variable called `high_payment_provider` is created. Suppliers are labeled as `1` if their standardized payment amount is above the 75th percentile (i.e., the top 25% of payment volume), and `0` otherwise. This framing turns the problem into a binary classification task: identifying whether a supplier is among the highest-paid providers.

- Finally, the distribution of the target variable is printed to verify the class balance.


```python
# Convert the payment column to numeric, just in case
df['Suplr_Mdcr_Stdzd_Pymt_Amt'] = pd.to_numeric(df['Suplr_Mdcr_Stdzd_Pymt_Amt'], errors='coerce')

# Drop rows with missing target values
df = df.dropna(subset=['Suplr_Mdcr_Stdzd_Pymt_Amt'])

# Create binary target: 1 if above the 75th percentile, 0 otherwise
threshold = df['Suplr_Mdcr_Stdzd_Pymt_Amt'].quantile(0.75)
df['high_payment_provider'] = (df['Suplr_Mdcr_Stdzd_Pymt_Amt'] > threshold).astype(int)

# Confirm distribution of target
print("Target distribution:")
print(df['high_payment_provider'].value_counts(normalize=True))
```

### 5. Feature Cleaning and Preprocessing

In this section, the dataset is refined by removing irrelevant, non-informative, or potentially data-leaking variables prior to model training. This includes:

- Dropping identifier columns (e.g., provider names, addresses, and zip codes) that do not contribute predictive value and may introduce noise.
- Removing columns directly related to the target variable, such as total charges or standardized payment amounts, to prevent label leakage.
- Eliminating features with excessive missing values or those unlikely to provide generalizable insight.
- Converting categorical variables into binary format using one-hot encoding to ensure compatibility with machine learning algorithms.

This step is essential to improve model robustness, reduce dimensionality, and prevent overfitting to provider-specific identifiers or redundant administrative fields.


```python
# Drop columns that are identifiers or directly related to the target
columns_to_remove = [
    'Suplr_NPI', 'Suplr_Prvdr_Last_Name_Org', 'Suplr_Prvdr_First_Name',
    'Suplr_Prvdr_MI', 'Suplr_Prvdr_Crdntls', 'Suplr_Prvdr_St1', 'Suplr_Prvdr_St2',
    'Suplr_Prvdr_City', 'Suplr_Prvdr_State_Abrvtn', 'Suplr_Prvdr_State_FIPS',
    'Suplr_Prvdr_Zip5', 'Suplr_Prvdr_Cntry', 'Suplr_Prvdr_Spclty_Desc',
    'Suplr_Prvdr_Spclty_Srce', 'Suplr_Prvdr_Ent_Cd', 'Suplr_Prvdr_RUCA_Desc',
    'Suplr_Prvdr_RUCA', 'DME_Sprsn_Ind', 'POS_Sprsn_Ind', 'Drug_Sprsn_Ind',
    'Suplr_Mdcr_Stdzd_Pymt_Amt', 'Suplr_Mdcr_Pymt_Amt', 'Suplr_Mdcr_Alowd_Amt',
    'Suplr_Sbmtd_Chrgs'
]

# Drop from DataFrame if they exist
df = df.drop(columns=[col for col in columns_to_remove if col in df.columns], errors='ignore')

# Drop rows with too many missing values
df = df.dropna(thresh=int(len(df.columns) * 0.7))

# Drop columns with too many missing values (>30%)
df = df.loc[:, df.isnull().mean() < 0.3]

# Convert all remaining object columns to categorical or drop them if not usable
df = df.select_dtypes(include=['float64', 'int64', 'object'])  # remove unnecessary types
df = pd.get_dummies(df, drop_first=True)

# Confirm shape after encoding
print("Shape after feature selection and encoding:", df.shape)
```

### 6. Model Training and Evaluation

This section focuses on building predictive models to classify high-payment Medicare suppliers based on their operational and patient profile characteristics.

#### 1. Feature and Target Definition
The feature matrix (`X`) is derived from all available numeric and encoded variables, excluding the target variable (`high_payment_provider`). Non-numeric fields are converted using one-hot encoding to ensure compatibility with scikit-learn classifiers.

#### 2. Data Splitting and Imputation
A stratified train-test split is used to preserve class distribution, with 80% of the data allocated for training. Since the dataset contains missing values, median imputation is applied to both training and test sets to ensure all models can operate on complete numerical data.

#### 3. Model Benchmarking
A diverse set of classifiers is compared, including:
- **Logistic Regression** (with class balancing)
- **Random Forest**
- **Gradient Boosting**
- **Decision Tree**
- **Naive Bayes**
- **Support Vector Machine (SVM)**

These models are selected to represent a variety of learning strategies (linear, tree-based, probabilistic, and kernel methods).

Each model is evaluated using the following metrics:
- **Accuracy**: Overall correctness of predictions.
- **Precision**: Correctness of positive predictions.
- **Recall**: Sensitivity to detecting high-payment providers.
- **F1 Score**: Harmonic mean of precision and recall.
- **ROC-AUC**: Discriminative ability across classification thresholds.

All results are compiled into a comparison table and sorted by ROC-AUC to identify the best-performing algorithm under imbalanced class conditions.


```python
# 1. Define features and target
X = pd.get_dummies(df.drop(columns=['high_payment_provider']), drop_first=True)
y = df['high_payment_provider']

# 2. Stratified train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# 3. Impute missing values (numerical strategy: median)
imputer = SimpleImputer(strategy='median')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# 4. Scale only for logistic regression
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_imputed)
X_test_scaled = scaler.transform(X_test_imputed)

# 5. Define models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, class_weight='balanced'),
    'Random Forest': RandomForestClassifier(random_state=42, class_weight='balanced'),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42, class_weight='balanced'),
    'Naive Bayes': GaussianNB(),
    'Support Vector Machine': SVC(probability=True, class_weight='balanced')
}

# 6. Evaluation loop
results = []
print("Starting model comparison...\n")

for name, model in models.items():
    print(f"Training and evaluating: {name}")
    
    # Use scaled data only for logistic regression
    if name == 'Logistic Regression':
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        y_proba = model.predict_proba(X_test_scaled)[:, 1]
    else:
        model.fit(X_train_imputed, y_train)
        y_pred = model.predict(X_test_imputed)
        y_proba = model.predict_proba(X_test_imputed)[:, 1] if hasattr(model, "predict_proba") else y_pred

    print(f"{name} - Training and prediction completed.\n")
    
    results.append({
        'Model': name,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1 Score': f1_score(y_test, y_pred),
        'ROC-AUC': roc_auc_score(y_test, y_proba)
    })

# 7. Results
results_df = pd.DataFrame(results).sort_values(by='ROC-AUC', ascending=False)
print("Comparison Table:\n")
print(results_df.round(4))
```

#### Key Observations

- **Gradient Boosting** delivered the best overall performance across all metrics, particularly in **F1 Score** and **ROC-AUC**, which are crucial in imbalanced classification tasks.
- **Random Forest** followed very closely, with near-identical performance metrics, indicating strong robustness and generalization.
- Both tree-based ensemble methods outperform Logistic Regression in terms of discriminative power (ROC-AUC), while maintaining high recall and precision.
- Support Vector Machine and Naive Bayes achieved extremely high precision, but their recall scores were significantly lower, leading to less balanced predictions.
- The standalone Decision Tree underperformed compared to its ensemble variants (Random Forest and Gradient Boosting), which is consistent with expectations regarding variance and overfitting.

#### Gradient Boosting vs. Random Forest

- **Gradient Boosting** builds trees sequentially, correcting the errors of prior models. This leads to a more finely tuned ensemble that typically performs better in complex patterns, as reflected in the slightly higher F1 and ROC-AUC scores.
- **Random Forest**, on the other hand, builds trees independently using bagging, which enhances robustness and reduces variance. It achieved similar results but may require less hyperparameter tuning and is often faster to train.
- In this dataset, Gradient Boosting’s marginal improvements suggest that it better captures subtle non-linearities and interactions across features.

**Conclusion**: While both models are strong candidates, **Gradient Boosting is the preferred choice** due to its superior overall balance between precision, recall, and ROC-AUC.


### 7. Gradient Boosting Model and Feature Interpretation

This section focuses on training a final classification model using the Gradient Boosting algorithm, which previously demonstrated the best performance across evaluation metrics.

Key steps include:
- Training the `GradientBoostingClassifier` on the training set
- Evaluating the model using standard classification metrics and ROC-AUC
- Extracting and visualizing the top 15 most influential features based on the model's internal importance scores

This analysis supports interpretability and helps identify the primary drivers of high Medicare payments across suppliers.


```python
# 1. Train the Gradient Boosting model
model = GradientBoostingClassifier(random_state=42)
model.fit(X_train_imputed, y_train)

# 2. Predict and evaluate
y_pred = model.predict(X_test_imputed)
y_proba = model.predict_proba(X_test_imputed)[:, 1]
roc_auc = roc_auc_score(y_test, y_proba)

print("Classification Report:")
print(classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print(f"\nROC-AUC Score: {roc_auc:.4f}")

# 3. Feature importance
importances = pd.Series(model.feature_importances_, index=X.columns)
top_features = importances.sort_values(ascending=False).head(15)

# 4. Plot top features
plt.figure(figsize=(8, 6))
top_features.plot(kind='barh')
plt.title("Top 15 Most Important Features (Gradient Boosting)")
plt.xlabel("Feature Importance")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# Print for reference
print("\nTop 15 Features:\n")
print(top_features)
```

### 8. Creative Scenario Simulation: Impact of Increasing Service Volume on High-Payment Classification

This section explores a hypothetical business scenario in which a Medicare equipment supplier increases the total number of services delivered. The objective is to assess whether such an operational change would influence the model's classification of that supplier as a "high payment" entity.

To achieve this, a borderline test case is selected from the model’s predictions: a supplier who was originally classified as **not high payment** but with a relatively high predicted probability (between 40% and 50%). The total number of services provided by this supplier is then tripled to simulate a significant expansion of operations.

The model is subsequently used to generate new predictions and probability scores for both the original and modified versions of the supplier data. This allows for a direct comparison of the model's behavior before and after the simulated intervention.

A visual representation is included in the form of a bar chart to clearly illustrate the change in predicted probability. This analysis provides insight into the sensitivity of the classification model to operational metrics such as service volume and supports interpretability for stakeholders evaluating cost and performance dynamics among Medicare equipment suppliers.


```python
# Creative Scenario (warning-free version): Tripling services for a borderline low-payment provider

# Reset test labels and input
y_test_reset = y_test.reset_index(drop=True)
X_test_df = pd.DataFrame(X_test_imputed, columns=X.columns)

# Combine predictions and probabilities for all test set
all_probs = model.predict_proba(X_test_df.values)[:, 1]

# Find a sample with relatively high probability but still classified as NOT high payment
borderline_index = np.where((all_probs > 0.4) & (all_probs < 0.5) & (y_test_reset == 0))[0]

if len(borderline_index) == 0:
    print(" No borderline cases found in test set.")
else:
    # Use the first such borderline case
    idx = borderline_index[0]
    original_sample_df = X_test_df.iloc[[idx]].copy()
    modified_sample_df = original_sample_df.copy()

    # Print current services
    print(f"Original 'Tot_Suplr_Srvcs' value: {original_sample_df['Tot_Suplr_Srvcs'].values[0]}")

    # Tripling services
    modified_sample_df['Tot_Suplr_Srvcs'] *= 3

    # Convert to arrays before prediction (to suppress warning)
    original_sample = original_sample_df.values
    modified_sample = modified_sample_df.values

    # Predictions
    original_pred = model.predict(original_sample)[0]
    original_proba = model.predict_proba(original_sample)[0][1]

    modified_pred = model.predict(modified_sample)[0]
    modified_proba = model.predict_proba(modified_sample)[0][1]

    # Display results
    print("\n--- Creative Scenario: Tripling Services for a Borderline Supplier ---")
    print(f"Original Prediction     → {'HIGH payment' if original_pred == 1 else 'NOT high payment'} (Prob: {original_proba:.4f})")
    print(f"Modified Prediction     → {'HIGH payment' if modified_pred == 1 else 'NOT high payment'} (Prob: {modified_proba:.4f})")

    # Interpretation
    if modified_pred != original_pred:
        print("\n The model CHANGED its prediction after tripling services.")
    elif modified_proba - original_proba > 0.1:
        print("\n Prediction remained the same, but probability increased significantly.")
    else:
        print("\n Prediction unchanged with minor probability shift.")


    # Visual representation: bar chart of prediction probabilities


    labels = ['Original Sample', 'Modified Sample (x3 Services)']
    probabilities = [original_proba, modified_proba]

    plt.figure(figsize=(6, 4))
    bars = plt.bar(labels, probabilities)
    plt.ylim(0, 1)
    plt.ylabel('Predicted Probability of High Payment')
    plt.title('Impact of Tripling Services on Prediction')

    # Add value labels
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height + 0.02, f'{height:.2f}', ha='center', va='bottom')

    plt.tight_layout()
    plt.show()
```
